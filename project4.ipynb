```python
import pandas as pd

# Load and preview
df = pd.read_csv('stock_data_july_2025.csv')
print(df.shape)
df.head()
```

```python
# Check data types and missing values
df.info()
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print("Unique tickers:", df['Ticker'].nunique())
print("Sectors:", df['Sector'].unique())
print("Date range:", df['Date'].min(), "to", df['Date'].max())
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Closing Prices
plt.figure(figsize=(12,6))
sns.histplot(df['Close Price'], bins=100)
plt.title('Distribution of Closing Prices')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Sector', y='PE Ratio', data=df)
plt.xticks(rotation=90)
plt.title('PE Ratio by Sector')
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='Sector', y='Dividend Yield', data=df)
plt.xticks(rotation=90)
plt.title('Dividend Yield by Sector')
plt.show()
```

```python
aapl = df[df['Ticker'] == 'AAPL']
plt.figure(figsize=(14,6))
plt.plot(aapl['Date'], aapl['Close Price'])
plt.title('AAPL Closing Price, June-July 2025')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
sector_avg = df.groupby('Sector')['Close Price'].mean().sort_values(ascending=False)
print(sector_avg)
```

```python
div_yield = df.groupby('Sector')['Dividend Yield'].mean().sort_values(ascending=False)
print(div_yield)
```

```python
df['Daily Return'] = df.groupby('Ticker')['Close Price'].pct_change()
```

```python
vol_df = df.groupby('Ticker')['Daily Return'].std()
vol_df = vol_df.sort_values(ascending=False)
print(vol_df.head(10))  # Volatile tickers
```

```python
top_eps = df.sort_values('EPS', ascending=False).head(10)
top_mcap = df.sort_values('Market Cap', ascending=False).head(10)
print(top_eps[['Ticker', 'EPS']])
print(top_mcap[['Ticker', 'Market Cap']])
```

```python
high_pe = df[df['PE Ratio'] > 100]
low_pe = df[df['PE Ratio'] < 5]
print(high_pe[['Ticker', 'PE Ratio']])
print(low_pe[['Ticker', 'PE Ratio']])
```

```python
df.to_csv('stock_data_july_2025_analysis_results.csv', index=False)
```

```python
for ticker in ['AAPL', 'MSFT', 'XOM', 'T']:
    subset = df[df['Ticker'] == ticker]
    plt.plot(pd.to_datetime(subset['Date']), subset['Close Price'], label=ticker)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Daily Closing Prices (AAPL, MSFT, XOM, T)')
plt.show()
```
